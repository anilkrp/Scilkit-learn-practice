```python
cat_feats = ['purpose']
```

```python
final_df = pd.get_dummies(df,columns=cat_feats,drop_first=True)
```

```python
final_df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = final_df.drop('not.fully.paid',axis=1)
y = final_df['not.fully.paid']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(x_train,y_train)
```

```python
pred = dtree.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc_prd = rfc.predict(x_test)
```

```python
print(classification_report(y_test,rfc_prd))
```

```python
print(confusion_matrix(y_test,rfc_prd))
```
